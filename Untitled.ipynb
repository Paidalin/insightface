```python
import face_model
import argparse
import cv2
import sys
import easydict
import numpy as np
```

```python
args = easydict.EasyDict({
    "image_size": '112,112',
    "model": 'model/model,0',
    "ga_model": 'model/model,0',
    "gpu": 0,
    "det": 0,
    "flip": 0,
    "threshold": 1.24
})
```

```python
# 调用摄像头，查看特征点
model = face_model.FaceModel(args)

point_size = 1
point_color = (0, 0, 255)
thickness = 4
cap = cv2.VideoCapture(0)
while (True):
    ret, frame = cap.read()
    try:
        img, points = model.get_input(frame)
    except:
        continue
    for point in points:
#         print(point)
        cv2.circle(frame, (point[0], point[1]), point_size, point_color, thickness)
    cv2.imshow('...', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print('done')
        break
cap.release()
cv2.destroyAllWindows()
```

```python
# 修改两层子目录的中文名称
import os
import pypinyin
from pypinyin import pinyin, lazy_pinyin
 
path='./OFD_full_DB_labelled'
filelist=os.listdir(path)#该文件夹下所有的文件（包括文件夹）
for files in filelist:#遍历所有文件
    path2 = path + '/' + files
    filelist2=os.listdir(path2)
    for files2 in filelist2:
        Olddir=os.path.join(path2,files2)#原来的文件路径
        filename=os.path.splitext(files2)[0] #获取文件名
        #把文件名里的汉字转换成其首字母
        filename1 = pinyin(filename, style=pypinyin.FIRST_LETTER) 
        #创建一个空列表
        filename2 = []
        for ch in filename1:
            filename2.extend(ch)
    #把列表转换成没有间隔的字符串，因为文件名要以字符串形式存在
        filenameToStr = ''.join(filename2)

        filetype=os.path.splitext(files2)[1]#文件扩展名

        Newdir=os.path.join(path2,filenameToStr + filetype)#新的文件名
        print(Olddir)
        print(Newdir)
        os.rename(Olddir,Newdir)#重命名
```
